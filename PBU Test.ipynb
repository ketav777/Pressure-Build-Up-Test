# Calculating the slope of the straight line part will be done once again when i have studied linear regression. DO NOT FORGET THIS TASK !!!

```python
# Considering the straight line portion of the Semi-Log Plot , I need to find its slope to determine the average permeability (k)
```

```python
# AND Remember , the graph is inverted about the X-AXIS.....so points that are closer to the left side of the straight portion are actually some of the last points in the excel sheet

m = - (3310 - 3306) / (np.log10(24.682200) - np.log10(31.845771))

# Now , Average Permeability can be easily calculated by using the formula : K = 162.6 * Bo * mu_o * Qo / m * H

K = (162.6 * Bo * mu_o * Qo) / (m * H)

print(f"Average Permeability is approximately equal to {K} milli-darcy")
```

```python
# Getting P1hr from the Semi_Log Plot

P1hr = 3266

# Skin Factor Calculation

S = 1.151 * ((P1hr - Pwf_initial) / m - np.log10(K / (phi * mu_o * ct * rw**2)) + 3.23)
print(f"The Skin Factor has a value of {S}")
```
